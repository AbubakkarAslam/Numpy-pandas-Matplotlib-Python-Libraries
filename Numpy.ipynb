```python
#stacking means to add the arrays in horizantoly and vertically and make dimensional array
```

```python
a = [1,2,3]
b = [4,5,6]
stack = np.vstack((a,b))
print(stack)
```

```python
stack = np.hstack((a,b))
print(stack)
```

```python
c = np.array([[1,2,3],[4,5,6]])
print(c)
```

```python
vsplit = np.vsplit(c,2)
print(vsplit)
for s in vsplit:
    print(s)
# splitting means to divide the array in equal parts in columns
```

```python
hsplit = np.hsplit(c,3)
print(hsplit)
for s in hsplit:
    print(s)
# splitting means to divide the array in equal parts in 3 columns
```
