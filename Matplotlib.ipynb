```python
import matplotlib.pyplot as plt
```

```python
#create first chart
x = [1,2,3,4,5,6,7,8]
y = [12,32,45,54,67,76,89,98]
plt.plot(x,y)
plt.show()
```

```python
#coustomize chart
plt.figure(figsize=(8,3)) # this is for setting the chart figure
plt.plot(x,y,color='green',marker='o',linestyle='--',linewidth='2',markersize=12)
plt.show()
```

```python
#for lable the chart
plt.figure(figsize=(8,3)) # this is for setting the chart figure
plt.plot(x,y,color='green',marker='o',linestyle='--',linewidth='2',markersize=12)
plt.title("Ya aik chart ha")
plt.xlabel("X-axis lable")
plt.ylabel("Y-axis labe")

plt.show()
```

```python
#Advanced line chart - multiple line & lagends
#shows multiple lines in the same chart by using the data again and again with different new other data 
x = [1,2,3,4,5,6,7,8]
y = [12,32,45,54,67,76,89,98]
z = [22,42,55,64,77,86,99,108]
plt.plot(x,y,label='Sales 2025')
plt.plot(x,z,label='Sales 2026')

plt.title('sales chart')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.legend() #show legends

plt.show()
```

```python
#bar chart
```

```python
x = ['A','B','C','D','E','F','G','H']
y = [12,32,105,54,67,76,89,98]
plt.bar(x,y)
plt.title('Bar chart')
plt.show()
```

```python
#5 Histogram : Used for distribution analysis
# Bar and Histogram chart are similar but not same :
#bar is catagical data with gap but histogram is continous data with not gap due to continous
```

```python
import random
```

```python
data = [random.randint(1,20) for _ in range(500)]
print(data)
```

```python
plt.hist(data,bins=20)
plt.title('Historam')
plt.show()
```

```python
#6 pie chart 
#is used to show the relationships part of the whole data set
# also used to show part-to-whole relationship/data
# it is collecction of complete 360
```

```python
catagories = ['A','B','C','D','E','F','G','H']
sales = [12,32,105,54,67,76,89,98]
plt.pie(sales, labels=catagories, autopct='%1.1f%%',startangle=90)
plt.title('Pie Chart')
plt.show()
```

```python
#scatter plot:used to find the relationships between variables
```

```python
y1 = [12,20,30,40,50,60]
y2 = [23,34,45,67,78,89]
plt.scatter(y1,y2)
plt.title('Scatter chart')
plt.show()
```

```python
#subplots: used to show multiple chatrs in one figure and also to set the position according to our need
```

```python
#data 1
x = ['A','B','C','D','E','F','G','H']
y = [12,32,105,54,67,76,89,98]

#data 2
y1 = [12,20,30,40,50,60]
y2 = [23,34,45,67,78,89]

plt.figure(figsize=(6,3))

plt.subplot(1,2,1)
plt.bar(x,y)
plt.title('Bar chart')

plt.subplot(1,2,2)
plt.scatter(y1,y2)
plt.title('scatter chart')

plt.show()
```
